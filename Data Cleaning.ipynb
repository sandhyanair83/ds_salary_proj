```python
# Removing records where Salary Estimate is -1
df = df[df['Salary Estimate']!= '-1']
```

```python
df
```

#### Salary Parsing

```python
#Add a column hourly in our dataframe df, which indicates if the salary given is hourly
df['hourly']= df['Salary Estimate'].apply(lambda x:1 if 'per hour' in x.lower() else 0)
```

```python
#Add a column employer_provided_sal in our dataframe df, which indicates if the salary is provided by the employer
df['employer_provided']= df['Salary Estimate'].apply(lambda x:1 if 'employer provided salary:' in x.lower() else 0)
```

```python
df.head(50)
```

```python
# Removing the text (Glassdoor est.) from the Salary Estimate field by applying the lambda function
Salary = df['Salary Estimate'].apply(lambda x:x.split('(')[0])
```

```python
Salary.head(50)
```

```python
# Removing the K and $ from the Salary by using lambda function
remove_kd = Salary.apply(lambda x:x.replace('K','').replace('$',''))
remove_kd.head(50)
```

```python
# replace per hour and employer provided salary text with space ''
min_hr = remove_kd.apply(lambda x:x.lower().replace('per hour','').replace('employer provided salary',''))
min_hr
```

```python
#Insert column for min salary and max salary
df['min_salary'] = min_hr.apply(lambda x: x.split('-')[0])
df['max_salary'] = min_hr.apply(lambda x: x.split('-')[1])
```

```python
df['min_salary']=df['min_salary'].apply(lambda x:x.replace(':',''))#replacing the :sign in front of some values with space
```

```python
df['min_salary']=df['min_salary'].astype(int)#converting min salary to int data type
```

```python
df['max_salary']=df['max_salary'].astype(int)#converting max salary to int data type
```

```python
df['avg_salary'] = (df['min_salary']+df['max_salary'])/2
```

```python
df
```

#### Extract only Company Name from the Company Name field

```python
df['company_txt'] = df.apply(lambda x:x['Company Name'] if x['Rating']<0 else x['Company Name'][:-4],axis=1)
```

```python
df
```

#### Extract the States from the Location field

```python
df['job_state'] = df['Location'].apply(lambda x:x.split(',')[1])
```

```python
df.job_state.value_counts()
```

```python
#Adding a new column same_state which will indicate 1 if Location and Headquaerters are same ,0 if different
df['same_state']= df.apply(lambda x: 1 if x.Location==x.Headquarters else 0, axis =1)
df
```

```python
#### Extract the age of the company
df['company_age'] = df['Founded'].apply(lambda x: x if x<1 else 2021-x)
```
